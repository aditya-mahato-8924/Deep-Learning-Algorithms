```python
class ANN:
  def __init__(self, epochs=2000, learning_rate=0.01, loss='binary_crossentropy', activation='sigmoid'):
    self.epochs = epochs
    self.learning_rate = learning_rate
    self.loss = loss
    self.parameters = None
    self.activation = activation

  def sigmoid(self, Z):
    return 1 / (1+np.exp(-Z))

  def initialize_parameters(self, layer_dim):
    parameters = {}

    for i in range(1, len(layer_dim)):
      parameters['W'+str(i-1)] = np.ones((layer_dim[i-1], layer_dim[i])) * 0.001
      parameters['b'+str(i-1)] = np.zeros((layer_dim[i], 1))

    return parameters

  def forward_propagation(self, x, layer_dim):
    ## previous layer activation, current layer activation
    A_prev, A_curr = x, x

    for i in range(len(layer_dim)-1):
      A_prev = A_curr
      W = self.parameters['W'+str(i)]
      b = self.parameters['b'+str(i)]

      Z = np.dot(W.T, A_prev) + b
      A_curr = self.sigmoid(Z)

    return A_prev, A_curr

  def binary_crossentropy(self, y_true, y_pred):
    epsilon = 1e-9
    return -(y_true*np.log(y_pred+epsilon) + (1-y_true)*np.log(1-y_pred+epsilon))

  def update_parameters(self, parameters, A_prev, A_curr, y, x):
    y_predict = A_curr[0][0]

    parameters['W1'][0][0] += self.learning_rate * (y-y_predict) * (A_prev[0][0])
    parameters['W1'][1][0] += self.learning_rate * (y-y_predict) * (A_prev[1][0])
    parameters['b1'][0][0] += self.learning_rate * (y-y_predict)

    parameters['W0'][0][0] += self.learning_rate * (y-y_predict) * (parameters['W1'][0][0]) * (A_prev[0][0]) * (1-(A_prev[0][0])) * x[0][0]
    parameters['W0'][1][0] += self.learning_rate * (y-y_predict) * (parameters['W1'][0][0]) * (A_prev[0][0]) * (1-(A_prev[0][0])) * x[1][0]
    parameters['b0'][0][0] += self.learning_rate * (y-y_predict) * (parameters['W1'][0][0]) * (A_prev[0][0]) * (1-(A_prev[0][0]))

    parameters['W0'][0][1] += self.learning_rate * (y-y_predict) * (parameters['W1'][1][0]) * (A_prev[1][0]) * (1- (A_prev[1][0])) * x[0][0]
    parameters['W0'][1][1] += self.learning_rate * (y-y_predict) * (parameters['W1'][1][0]) * (A_prev[1][0]) * (1- (A_prev[1][0])) * x[1][0]
    parameters['b0'][1][0] += self.learning_rate * (y-y_predict) * (parameters['W1'][1][0]) * (A_prev[1][0]) * (1- (A_prev[1][0]))

    return parameters

  def fit(self, X_train, y_train):
    ## we are implementing a simple neural network of 1 input layer of 2 nodes,
    ## 1 hidden layer of 2 nodes and 1 output layer of 1 node and activation function of all nodes is linear
    self.layer_dim = [2, 2, 1]

    ## initialize parameters i.e assign values to weights as 1 and bias as 0
    self.parameters = self.initialize_parameters(self.layer_dim)

    ## running a loop epochs number of times
    for i in range(self.epochs):
      ## creating empty list to store loss at each epoch
      Loss = []

      ## pick a data point from X_train
      for idx in range(len(X_train)):
        x = X_train[idx]
        y = y_train[idx]

        ## make prediction of that data point using forward propagation
        A_prev, A_curr = self.forward_propagation(x, self.layer_dim)
        y_predict = A_curr[0][0]

        ## calculate loss
        loss = self.binary_crossentropy(y, y_predict)

        ## update parameters
        self.parameters = self.update_parameters(self.parameters, A_prev, A_curr, y, x.reshape(2, 1))

        Loss.append(loss)

      # print(f"Epoch {i+1} : {np.mean(Loss)}")

  def predict(self, X_test):
    y_prob = np.zeros((X_test.shape[0], 2))

    for idx in range(X_test.shape[0]):
      x = X_test[idx]
      A_prev, A_curr = self.forward_propagation(x, self.layer_dim)
      y_prob[idx][0] = A_curr[0][0] ## probability of class to be 1
      y_prob[idx][1] = 1 - A_curr[0][0] ## probability of class to be 0

    print(y_prob)
    return np.where(y_prob[:, 0] > 0.5, 1, 0)
```
